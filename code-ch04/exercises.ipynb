```python

from ecc import G, S256Point, Signature, encode_base58, PrivateKey
```

```python
p1 = 5000 * G
p2 = (2018 ** 5) * G
p3 = 0xdeadbeef12345 * G

"""
this is equivalen to:
p1 = PrivateKey(5000)
p2 = PrivateKey((2018 ** 5))
etc.
"""

print(p1.sec(compressed=False).hex())

print(p2.sec(compressed=False).hex())

print(p3.sec(compressed=False).hex())
```

```python
print(p1.sec().hex())

print(p2.sec().hex())

print(p3.sec().hex())
```
